```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

LINE-PLOT

```python
data = {'day':[1,2,3,4,5], 'sales':[10,15,7,12,9]}
sns.lineplot(x='day', y='sales', data=data, marker='o', color='green')
plt.title("Line Plot")
plt.show()
```

BAR CHART

```python
df = pd.DataFrame({'category': ['A', 'B', 'C'], 'value': [5,7,3]})
sns.barplot(x='category', y='value', data=df, palette='pastel')
plt.title("Bar Chart")
plt.show()
```

SCATTER PLOT

```python
df = pd.DataFrame({'x':[5,7,8,7,2,4,9], 'y':[99,86,87,88,100,86,103], 'group':['A','B','A','B','A','B','A']})
sns.scatterplot(x='x', y='y', data=df, hue='group', s=100, palette='bright')
plt.title("Scatter Plot")
plt.show()
```

HISTOGRAM

```python
data = [1,2,2,3,3,3,4,4,5,5,5,5]
sns.histplot(data, bins=5, color='teal', kde=True) 
plt.title("Histogram")
plt.show()
```

BOX PLOT

```python
data = [20, 22, 23, 24, 25, 30, 35, 100]
sns.boxplot(y=data, color="lightblue")
plt.title("Box Plot")
plt.show()
```

VIOLIN PLOT

```python
data = [20, 22, 23, 24, 25, 30, 35]

sns.violinplot(y=data, inner="quartile", color="lightgreen")
plt.title("Violin Plot")
plt.show()
```

HEATMAP

```python
data = np.array([[1, 0.8, 0.3], [0.8, 1, 0.5], [0.3, 0.5, 1]])
df = pd.DataFrame(data, columns=['A','B','C'], index=['A','B','C'])

sns.heatmap(df, annot=True, cmap="coolwarm")
plt.title("Heatmap")
plt.show()
```

LABELS AND TITLE

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 12, 9]

sns.lineplot(x=x, y=y)
plt.title("Sales Trend", fontsize=15)
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

LEGENDS

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 15, 7, 12, 9]
y2 = [12, 15, 16, 14, 13]

sns.lineplot(x=x, y=y1, label="Product A")
sns.lineplot(x=x, y=y2, label="Product B")
plt.title("Sales", fontsize=12)
plt.legend(title="Products", loc="upper right")
plt.show()
```

STYLES AND THEMES

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 15, 7, 12, 9]
y2 = [12, 15, 16, 14, 13]

sns.lineplot(x=x, y=y1, label="Product A")
sns.lineplot(x=x, y=y2, label="Product B")
plt.title("Sales", fontsize=12)
plt.legend(title="Products", loc="upper right")
sns.set_style("white")
plt.show()
```

FIGURE SIZE AND LAYOUT AND GRIDS

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 15, 7, 12, 9]
y2 = [12, 15, 16, 14, 13]

plt.figure(figsize=(7,5))
sns.lineplot(x=x, y=y1, label="Product A")
sns.lineplot(x=x, y=y2, label="Product B")
plt.title("Sales", fontsize=12)
plt.legend(title="Products", loc="upper right")
plt.grid(True)
plt.show()
```
