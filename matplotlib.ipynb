```python
# import libraries
import matplotlib.pyplot as plt
```

LINE-PLOT

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 12, 9]

plt.plot(x, y, color='blue', linestyle='--', marker='o')
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

BAR CHART

```python
categories = ['A', 'B', 'C']
values = [5, 7, 3]

plt.bar(categories, values, color='blue')
plt.title("Bar Chart")
plt.xlabel("Category")
plt.ylabel("Values")
plt.show()
```

SCATTER PLOT

```python
x = [5,7,8,7,2,4,9]
y = [99,86,87,88,100,86,103]

plt.scatter(x, y, color='red', s=100) 
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

HISTOGRAM

```python
data = [1,2,2,3,3,3,4,4,5,5,5,5]

plt.hist(data, bins=5, color='purple', alpha=0.7)
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

BOX PLOT

```python
data = [20, 22, 23, 24, 25, 30, 35, 100]

plt.boxplot(data)
plt.title("Box Plot")
plt.ylabel("Values")
plt.show()
```

VIOLIN PLOT

```python
data = [20, 22, 23, 24, 25, 30, 35]

plt.violinplot(data, showmeans=True)
plt.title("Violin Plot")
plt.ylabel("Values")
plt.show()
```

HEATMAP

```python
import numpy as np

data = np.array([[1, 0.8, 0.3], [0.8, 1, 0.5], [0.3, 0.5, 1]])

plt.imshow(data)
plt.colorbar()
plt.title("Heatmap")
plt.show()
```

LABELS AND TITLE

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 12, 9]

plt.plot(x, y)
plt.title("Sales Trend", fontsize = 15)
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

LEGENDS

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 15, 7, 12, 9]
y2 = [12, 15, 16, 14, 13]

plt.plot(x, y1, label="Product A")
plt.plot(x, y2, label="Product B")
plt.title("Sales", fontsize=12)
plt.legend(title="Products", loc="upper right")
plt.show()
```

STYLES AND THEMES

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 15, 7, 12, 9]
y2 = [12, 15, 16, 14, 13]

plt.plot(x, y1, label="Product A")
plt.plot(x, y2, label="Product B")
plt.title("Sales", fontsize=12)
plt.legend(title="Products", loc="upper right")
plt.style.use("classic")
plt.show()
```

FIGURE SIZE AND LAYOUT AND GRIDS

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 15, 7, 12, 9]
y2 = [12, 15, 16, 14, 13]

plt.figure(figsize=(7,5))
plt.plot(x, y1, label="Product A")
plt.plot(x, y2, label="Product B")
plt.title("Sales", fontsize=12)
plt.legend(title="Products", loc="upper right")
plt.grid(True)
plt.show()
```

plt.subplot()

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 15, 7, 12, 9]
y2 = [12, 15, 16, 14, 13]

plt.subplot(1,2,1)
plt.plot(x, y1)
plt.title("Product A", fontsize=5)

plt.subplot(1,2,2)
plt.plot(x, y2)
plt.title("Product B", fontsize=5)
plt.show()
```

plt.subplots()

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 15, 7, 12, 9]
y2 = [12, 15, 16, 14, 13]

fig, axes = plt.subplots(1,2, figsize=(10,5))
axes[0].plot(x, y1)
axes[0].set_title("Product A", fontsize=5)

axes[1].plot(x, y2)
axes[1].set_title("Product B", fontsize=5)
plt.show()
```

tight_layout()

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 15, 7, 12, 9]
y2 = [12, 15, 16, 14, 13]

plt.subplot(1,2,1)
plt.plot(x, y1)
plt.title("Product A", fontsize=5)

plt.subplot(1,2,2)
plt.plot(x, y2)
plt.title("Product B", fontsize=5)

plt.tight_layout()
plt.show()
```

subplots_adjust()

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 15, 7, 12, 9]
y2 = [12, 15, 16, 14, 13]

plt.subplot(1,2,1)
plt.plot(x, y1)
plt.title("Product A", fontsize=5)

plt.subplot(1,2,2)
plt.plot(x, y2)
plt.title("Product B", fontsize=5)

plt.subplots_adjust(wspace=1, hspace=0.4)
plt.show()
```

sharing axes

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 15, 7, 12, 9]
y2 = [12, 15, 16, 14, 13]

fig, axes = plt.subplots(1,2, sharey=True)
axes[0].plot(x, y1)
axes[0].set_title("Product A", fontsize=5)

axes[1].plot(x, y2)
axes[1].set_title("Product B", fontsize=5)
plt.show()
```
