```python
import pandas as pd
```

PANDAS SERIES

```python
# creating series with index labels
a = [1,2,3,4,5]
series = pd.Series(a)
print(series)
```

```python
# using index to access elements of series
print(series[3])
```

```python
# creating labels
a = [1,3,5]
ser = pd.Series(a, index=["a", "b", "c"])
print(ser)
```

```python
# dictionary for series
calories = {"day1": 420, "day2": 380, "day3": 390}
ser = pd.Series(calories)
print(ser)
```

```python
calories = {"day1": 420, "day2": 380, "day3": 390}
ser = pd.Series(calories, index = ["day1", "day3"])
print(ser)
```

PANDAS DATAFRAME

```python
# creating dataframes
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}
df = pd.DataFrame(data)
print(df)
```

```python
# locating rows
print(df.loc[1])
```

```python
print(df.loc[[0,2]])
```

```python
df = pd.DataFrame(data, index = ["day1", "day2", "day3"])
print(df)
```

```python
print(df.loc["day3"])
```

WORKING WITH CSV DATASET

```python
df = pd.read_csv("student_depression_dataset.csv")
print(df.head(5))
```

```python
# data selection using loc
df.loc[4]       # row
```

```python
df.loc[[4,10]]            # multiple rows
```

```python
df.loc[5:9, ["CGPA", "Academic Pressure"]]
```

```python
df.loc[df["Age"]<25, ["Profession", "Depression"]]
```

```python
# data selection using iloc
df.iloc[3]
```

```python
df.iloc[34:45, 0:6]         # [rows, columns]
```

```python
df.iloc[[32,48,64], [2,4,5]]
```

```python
df[df["Age"]<20]
```

```python
df[(df["Age"]<20) & (df["City"] == "Bangalore")]
```

```python
df.query("Age<20 and City == 'Ludhiana'")
```

```python
# boolean indexing
mask = df["CGPA"] > 8.5
df[mask]
```

```python
df.loc[df["Depression"] == 1, ["Age", "City", "Depression"]]
```

```python
df[~(df["City"] == "Patna")]
```

```python
df[df["City"].isin(["Ludhiana", "Bangalore"])]
```

```python
df[df["Age"].between(20, 25)]
```

```python
# string based filtering
df[df["Gender"].str.startswith("F")]
```

```python
df[df["City"].str.contains("C")]
```
