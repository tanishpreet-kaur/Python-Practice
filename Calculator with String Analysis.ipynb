```python
def calculator():
    a = float(input("Enter first number: "))
    b = float(input("Enter second number: "))

    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")

    choice = input("Choose operation: ")

    if choice == "1":
        print("Result:", a + b)
    elif choice == "2":
        print("Result:", a - b)
    elif choice == "3":
        print("Result:", a * b)
    elif choice == "4":
        if b != 0:
            print("Result:", a / b)
        else:
            print("Cannot divide by zero")
    else:
        print("Invalid choice")
```

```python
def word_count():
    text = input("Enter a sentence: ")
    words = text.split()
    print("Word count:", len(words))
```

```python
def palindrome_check():
    text = input("Enter a word: ")
    if text == text[::-1]:
        print("It is a Palindrome")
    else:
        print("Not a palindrome")
```

```python
while True:
    print("\n------ MENU ------")
    print("1. Calculator")
    print("2. Word Count")
    print("3. Palindrome Check")
    print("4. Exit")

    option = input("Enter your choice: ")

    if option == "1":
        calculator()
    elif option == "2":
        word_count()
    elif option == "3":
        palindrome_check()
    elif option == "4":
        print("Exiting program......")
        break
    else:
        print("Invalid option")
```
