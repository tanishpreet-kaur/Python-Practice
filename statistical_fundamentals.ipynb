```python
# import libraries
import numpy as np
import pandas as pd
```

```python
data = [23,78,56,34,13,87,45,34,84,56,23,48,30,78,93,48, 23]
```

```python
# descriptive statistics using np
data_np = np.array(data)

print("Mean: ", np.mean(data_np))
print("Median: ", np.median(data_np))
print("Variance: ", np.var(data_np))
print("Standard Deviation: ", np.std(data_np))
```

```python
# if sample population
print("Sample Variance:", np.var(data_np, ddof=1))
print("Sample Std Dev:", np.std(data_np, ddof=1))
```

```python
# descriptive statistics using pandas
df = pd.DataFrame(data)
df.describe()
```

```python
# we need to calculate mode separately
print("Mode:", df.mode())
```
