```python
# import libraries
import pandas as pd
import numpy as np
```

DATA CLEANING

```python
df = pd.DataFrame({
    "Name": [" Alice ", "Bob", "Charlie", "Bob"],
    "Age": [20, np.nan, 22 , 20],
    "City": ["Delhi", "delhi", None, "Delhi"],
    "CGPA": [8.5, 9.1, None, 8.5]
})
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
```

```python
df["City"] = df["City"].fillna("Unknown")
```

```python
df["CGPA"] = df["CGPA"].interpolate()
```

```python
df.head()
```

```python
df.duplicated()
```

```python
df.drop_duplicates()
```

```python
df["Age"] = pd.to_numeric(df["Age"], errors = "coerce")
```

```python
df["Name"] = df["Name"].str.strip()
```

```python
df["City"] = df["City"].str.title()
```

```python
df["City"] = df["City"].replace("Unknown", "Delhi")
```

```python
df.head()
```

DATA TRANSFORMATION

```python
# log-transformation
df["CGPA_log"] = np.log(df["CGPA"])
```

```python
# reciprocal transformation
df["CGPA_reciprocal"] = 1 / (df["CGPA"])
```

```python
# square transformation
df["CGPA_square"] = (df["CGPA"]) ** 2
```

```python
# square-root transformation
df["CGPA_sqrt"] = (df["CGPA"]) ** 0.5
```
