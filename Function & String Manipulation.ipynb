FUNCTION DEFINITION & CALLING

```python
# basic function
def greet():
    print("Hello, World!")
    
greet()
```

```python
# function with parameter
def greet(name):
    print("Hello,", name)

greet("Tanish")
```

```python
# function with return value
def add(a, b):
    return a + b

result = add(5, 3)
print(result)
```

```python
# function that returns multiple values
def calculate(a, b):
    return a + b, a - b, a * b

add, diff, prod = calculate(10, 5)
print(add, diff, prod)
```

```python
# function calling another function
def square(x):
    return x * x

def cube(x):
    return square(x) * x

print(cube(3))
```

PARAMETERS

```python
# positional arguments
def power(a, b):
    return a ** b

print(power(2, 3))
```

```python
# keyword arguments
print(power(b=3, a=2))
```

```python
# default parameters
def greet(name="Guest"):
    print("Hello", name)

greet()
greet("Tanish")
```

```python
# variable length arguments (*args)
def total(*numbers):
    return sum(numbers)

print(total(1, 2, 3, 4))
```

```python
# keyword variable length arguments (*kwargs)
def student_info(**details):
    for key, value in details.items():
        print(key, ":", value)

student_info(name="Tanish", age=21, course="Python")
```

RETURN STATEMENT

```python
# return without value
def show():
    print("Displayed")
    return

show()
```

```python
# conditional return
def check_even(n):
    if n % 2 == 0:
        return "Even"
    return "Odd"

print(check_even(7))
```

SCOPE OF VARIABLES

```python
# local scope
def func():
    x = 10
    print(x)

func()
```

```python
# global scope
x = 20
def show():
    print(x)

show()
```

```python
# global keyword
count = 0

def increment():
    global count
    count += 1

increment()
print(count)
```

```python
# non-local scope
def outer():
    x = 5
    def inner():
        nonlocal x
        x += 1
    inner()
    print(x)

outer()
```

STRING OPERATIONS

```python
# string creation
s1 = "Python"
s2 = 'Programming'
s3 = """Multiline String"""
```

```python
# indexing
s = "Python"
print(s[0])
print(s[-1])
```

```python
# slicing
print(s[0:4])
print(s[:3])
print(s[::2])
print(s[::-1])
```

```python
# string concatenation
a = "Data"
b = "Science"
print(a + " " + b)
```

```python
# string repetition
print("lala" * 5)
```

```python
# membership operators
print("Py" in "Python")
print("Java" not in "Python")
```

```python
# case-conversion
text = "Tanish"

print(text.upper())
print(text.lower())
print(text.title())
print(text.capitalize())
```

```python
# whitespace removal
msg = "  hello  "

print(msg.strip())
print(msg.lstrip())
print(msg.rstrip())
```

```python
# search and replace
sentence = "I love Python"

print(sentence.find("Python"))
print(sentence.replace("Python", "Coding"))
```

```python
# splitting and joining
words = "Python is easy".split()
print(words)

joined = "-".join(words)
print(joined)
```

```python
# checking string properties
print("abc".isalpha())
print("123".isdigit())
print("abc123".isalnum())
```

```python
# string formatting using format()
name = "Tanish"
age = 21

print("My name is {} and I am {}".format(name, age))
```

```python
# indexed formatting
print("Age: {1}, Name: {0}".format(name, age))
```

```python
# named formatting
print("Name: {n}, Age: {a}".format(n=name, a=age))
```

```python
# f-strings
print(f"My name is {name} and I am {age}")
```

```python
# formatting numbers
pi = 3.14159
print(f"{pi:.2f}")
```
