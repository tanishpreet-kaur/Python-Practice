CONTROL-FLOW STATEMENTS

```python
# if statement
age = 18
if age >= 18:
    print("You are eligible to vote")
```

```python
# if-else statement
number = 7
if number % 2 == 0:
    print("Even number")
else:
    print("Odd number")
```

```python
# if-elif-else statement
marks = 85
if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 60:
    print("Grade C")
else:
    print("Fail")
```

```python
# nested if statement
age = 20
has_id = True

if age >= 18:
    if has_id:
        print("Entry allowed")
    else:
        print("ID required")
else:
    print("Underage")
```

```python
# if-else using expressions
a, b = 5, 10
max_value = a if a > b else b
print(max_value)
```

LOOPS

```python
# for loop without range
for i in range(5):
    print(i)
```

```python
# for loop with range
for i in range(1, 10, 2):
    print(i)
```

```python
# for loop to iterate over list
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```

```python
# nested for loop
for i in range(3):
    for j in range(2):
        print(i, j)
```

```python
# while loop
count = 1
while count <= 5:
    print(count)
    count += 1
```

```python
# while loop with condition
while True:
    user_input = input("Type 'exit' to stop: ")
    if user_input == "exit":
        break
```

```python
# nested while loop
i = 1
while i <= 3:
    j = 1
    while j <= 2:
        print(i, j)
        j += 1
    i += 1
```

```python
# break statement
for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
# continue statement
for i in range(5):
    if i == 2:
        continue
    print(i)
```

```python
# pass statement
for i in range(3):
    if i == 1:
        pass
    print(i)
```

PATTERN-PRINTING

```python
n = 5
for i in range(1, n + 1):
    print("*" * i)
```

```python
n = 5
for i in range(n, 0, -1):
    print("*" * i)
```

```python
n = 5
for i in range(1, n + 1):
    print(" " * (n - i) + "*" * (2 * i - 1))
```

```python
n = 4
for i in range(1, n + 1):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
```
