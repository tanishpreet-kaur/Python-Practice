```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import dataset
df = pd.read_csv("spotify_data.csv")
```

```python
# display dataset
df.head()
```

```python
# dataset statistics
df.info()
```

```python
df.describe()
```

## DATA CLEANING

```python
# check for missing values
df.isna().sum()
```

```python
# remove rows with missing value
df = df.dropna()
```

```python
# remove unnecessary columns
df = df.drop(columns=["Unnamed: 0", "track_id", "time_signature"])
```

```python
# display column names
df.columns
```

## DATA VISUALIZATION

```python
# barchart of track_genres
genre_counts = df['track_genre'].value_counts()
x = genre_counts.index
y = genre_counts.values

plt.figure(figsize=(15,14))
sns.barplot(x=x, y=y, palette='pastel')
plt.xlabel("Track Genres")
plt.ylabel("Song Count")
plt.title("Popularity of Songs based on Genres")
plt.xticks(rotation = 90)
plt.show()
```

```python
# check where there are any outliers
sns.boxplot(x=df['tempo'], color='lightblue')
```
